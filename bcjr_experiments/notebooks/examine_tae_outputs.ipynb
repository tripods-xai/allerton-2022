```python
bin_inputs = enumerate_binary_inputs(WINDOW).numpy()
```

```python
main_outputs = tae_outputs[:, WINDOW // 2 + 1, :]
```

```python
result = np.empty((WINDOW, WINDOW, 3))
for i in range(WINDOW):
    # bin_inputs is 2^W x W
    inputs_i_0 = bin_inputs[(bin_inputs[:, i] == 0)]  # Which of the 2^W is bit i == 0
    inds_i_0 = bitarray2dec(inputs_i_0, little_endian=False).numpy()
    assert np.all(inds_i_0 == np.arange(2 ** WINDOW)[(bin_inputs[:, i] == 0)])
    
    inputs_i_1 = np.array(inputs_i_0, copy=True)
    inputs_i_1[:, i] += 1
    inds_i_1 = bitarray2dec(inputs_i_1, little_endian=False).numpy()
    result[i] = np.mean((tae_outputs[inds_i_0] != tae_outputs[inds_i_1]).astype(np.float32), axis=0)
result
    
```
